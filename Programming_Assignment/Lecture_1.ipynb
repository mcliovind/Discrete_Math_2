*2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}*

$\textbf{Answer for num 2}$
--

**Step 1:** Let's prove Minkowski's inequality for sums by leveraging Hölder's inequality. Hölder's inequality states that for any two sequences of real numbers $ (x_k) $ and $ (y_k) $ and any $ p > 1 $ such that $ \frac{1}{p} + \frac{1}{q} = 1 $, the following holds:

$ \sum_{k=1}^n |x_k y_k| \leq \left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |y_k|^q \right)^{\frac{1}{q}}$

**Step 2:** Now, let's assign specific values to $ x_k $ and $ y_k $ as follows:

$
x_k = \frac{|a_k|^p}{\left(\sum_{k=1}^n |a_k|^p\right)^{\frac{1}{p}}}, \quad y_k = \frac{|b_k|^p}{\left(\sum_{k=1}^n |b_k|^p\right)^{\frac{1}{p}}}
$

**Step 3:** Substituting these expressions into the inequality, we obtain:

$
\sum_{k=1}^n |a_k + b_k|^p \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}} n^{\frac{1}{p-1}}
$

**Step 4:** Dividing both sides by $ n^{\frac{1}{p-1}} $, we get:

$
\left( \frac{1}{n^{\frac{1}{p-1}}} \sum_{k=1}^n |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}
$

**Step 5:** Taking the $ p $-th root on both sides, we conclude:

$
\left( \frac{1}{n} \sum_{k=1}^n |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}
$

Thus, Minkowski's inequality for sums is proven. **QED**

**Case 2:** Suppose either $ x $ or $ y $ is negative. Without loss of generality, assume $ x < 0 $. In this case, $ |x| = -x $ and $ |y| = -y $. Hence, $ |x+y| = -(x+y) $, which simplifies to $ |x+y| = -x - y $. Similarly, $ |x| + |y| = -x - y $. Since $ -(x+y) = -x - y $, the inequality remains valid in this case as well.

**Step 3:** Consequently, in all scenarios, $ |x+y| \leq |x| + |y| $ for all $ (x,y) $ in $ \mathbb{R} $.**QED**
