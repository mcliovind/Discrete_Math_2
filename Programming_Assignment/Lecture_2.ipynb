Task 1:
Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$



*Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$*

$\textbf{Task 1: Q1}$
-

We aim to demonstrate that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$ can be expressed in terms of the joint probability distribution $P(y)$.

**Step 1:** Definition of Expected Value

The expected value $E[y]$ is the sum of all possible values of $y$ weighted by their probabilities:

$$ E[y] = \sum_{\text{all } y} y \cdot P(y) $$

**Step 2:** Transformation of $E[y]$

We rewrite $E[y]$ using the properties of expectation:

$$ E[y] = \sum_{i=1}^n E[Y_i] $$

where $Y_i$ represents the number of trials for which the outcome falls into cell $i$.

**Step 3:** Individual Expected Values

For each $Y_i$, its expected value $E[Y_i]$ can be expressed as:

$$ E[Y_i] = \sum_{k=0}^n k \cdot P(Y_i = k) $$

**Step 4:** Probability of $Y_i$

Given the joint probability distribution:

$$ P(Y_i = k) = \frac{n!}{k!(n-k)!} p_i^k (1 - p_i)^{n-k} $$

**Step 5:** Substitute Into $E[Y_i]$

Substituting $P(Y_i = k)$ into $E[Y_i]$, we get:

$$ E[Y_i] = \sum_{k=0}^n k \cdot \frac{n!}{k!(n-k)!} p_i^k (1 - p_i)^{n-k} $$

**Step 6:** Simplification

After simplification, we find:

$$ E[Y_i] = np_i $$

**Step 7:** Final Conclusion

Since $E[y] = \sum_{i=1}^n E[Y_i]$, and each $E[Y_i] = np_i$, we conclude:

$$ E[y] = n \cdot p_i $$

Thus, the proof is complete.**QED.**

4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.
   <br><br>

*Question 1: From the following steps shown above, derive the probability distribution of the experiment.*

$\textbf{Task 2: Q1}$
-

The probability distribution of $Y$ follows the binomial distribution, which can be calculated using the binomial probability mass function:

$$ P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y} $$

**Step 1:** Explanation of Terms

- $P(Y = y)$ is the probability of observing $y$ successes in $n$ trials.
- $\binom{n}{y}$ is the binomial coefficient, representing the number of ways to choose $y$ successes out of $n$ trials.
- $p^y$ represents the probability of having $y$ successes.
- $q^{n-y}$ represents the probability of having $(n - y)$ failures. **QED**.

*Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.*

**Step 1:** Definition of Expected Value

The expected value $E[Y]$ is defined as:

$$ E[Y] = \sum_{\text{all } y} y \cdot P(y) $$

**Step 2:** Substitution of Binomial Distribution Formula

Substituting the binomial distribution formula $P(y)$ into $E[Y]$, we get:

$$ E[Y] = \sum_{\text{all } y} y \cdot \binom{n}{y} p^y (1-p)^{n-y} $$

**Step 3:** Simplification

Expanding and simplifying this expression, we find:

$$ E[Y] = np \sum_{y=0}^{n} \binom{n}{y} p^y (1-p)^{n-y} $$

**Step 4:** Application of Binomial Theorem

By the binomial theorem, the sum $\sum_{y=0}^{n} \binom{n}{y} p^y (1-p)^{n-y}$ equals $(p + (1-p))^n$, which is equal to $1^n = 1$.

**Step 5:** Conclusion

Hence, we conclude that $E[Y] = np$. **QED**.

Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

$\textbf{Task 2: Q4}$
-
**Given:**

- Number of trials, $n = 5$ (sampling 5 fuses)
- Probability of success (defective fuse), $p = 0.05$
- Probability of failure (non-defective fuse), $q = 1 - p = 1 - 0.05 = 0.95$

**Step 1:**
Let $Y$ represent the number of defective fuses observed during the 5 trials.

We aim to determine the probability $P(Y \geq 1)$, indicating the likelihood of observing at least one defective fuse.

**Step 2:**
By the complement rule, we find this probability as:

$$
P(Y \geq 1) = 1 - P(Y = 0)
$$

**Step 3:**
To find $P(Y = 0)$, the probability of observing no defective fuses, we apply the binomial probability formula:

$$
P(Y = k) = \binom{n}{k} \times p^k \times q^{n-k}
$$

**Step 4:**
Here, $k = 0$ (no defective fuses), $n = 5$, $p = 0.05$, and $q = 0.95$.

**Step 5:**
Let's compute $P(Y \geq 1)$:

$$
P(Y \geq 1) = 1 - P(Y = 0) \\
= 1 - \binom{5}{0} \times 0.05^0 \times 0.95^5 \\
= 1 - 1 \times 1 \times (0.95)^5 \\
= 1 - 1 \times 1 \times 0.7737809375 \\
\approx 1 - 0.7737809375 \\
\approx 0.2262190625
$$

**Step 6:**
Therefore, the probability of observing at least one defective fuse in a sample of 5 is approximately 0.2262 or 22.62%. **QED**

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

*Question 1: Find a general formula for the expected value of this distribution.*

$\textbf{Task 3: Q1}$
-
Consider a probability distribution of $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $.

**Step 1:**
To find a general formula for the expected value of this distribution, we calculate the sum of all possible outcomes weighted by their probabilities. The expected value $ E[y] $ is given by:
$$ E[y] = \sum_{y=0}^{\infty} y \cdot p(y) $$

**Step 2:**
Substituting the given probability distribution:
$$ E[y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $$

**Step 3:**
This expression represents the expected value of a Poisson distribution, where $ \lambda $ is the mean or expected value. Hence, we have:
$$ E[y] = \lambda $$

**Conclusion:**
The general formula for the expected value of this distribution is $ E[y] = \lambda $. **QED**

If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.*

-
Poisson distribution, expressed by the formula:

$$ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $$

Where:
- $\lambda $ represents the average rate, which in this case, is the mean density of seedlings per square yard.
- $ y $ denotes the number of seedlings in a region.

**Step 1: Assign Values**
Given that the mean density of seedlings is five per square yard, we have $\lambda = 5$.

$$ p(0) = \frac{5^0}{0!} e^{-5} = e^{-5} $$
